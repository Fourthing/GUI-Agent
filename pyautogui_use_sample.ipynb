 0.检验安装

```python
import pyautogui
print(pyautogui.position()) # 显示当前鼠标位置
```

1.鼠标操作

```python
import pyautogui
import time

# 鼠标移动到屏幕的中心
pyautogui.moveTo(960, 540, duration=1)  # 假设屏幕分辨率是1920x1080

# 鼠标点击
pyautogui.click()

# 鼠标按住并拖动
pyautogui.dragTo(100, 100, duration=1)

# 鼠标右键点击
pyautogui.rightClick()

# 鼠标滚轮操作
pyautogui.scroll(200)  # 向上滚动
pyautogui.scroll(-200)  # 向下滚动
```

2.键盘操作

```python
# 按键输入
pyautogui.write('Hello, this is an automated input!', interval=0.1)  # 每个字符输入的间隔为0.1秒

# 按下单个键
pyautogui.press('enter')

# 组合键（例如：Ctrl+C）
pyautogui.hotkey('ctrl', 'c')

# 键盘的特殊操作（例如：shift, tab）
pyautogui.keyDown('shift')
pyautogui.press('a')
pyautogui.keyUp('shift')
```

3.屏幕截图

```python
# 截取全屏截图并保存
screenshot = pyautogui.screenshot()
screenshot.save('screenshot.png')

# 截取屏幕的一部分
screenshot_part = pyautogui.screenshot(region=(0, 0, 300, 300))  # 左上角(0, 0)，宽300，高300
screenshot_part.save('screenshot_part.png')
```

4.屏幕位置与坐标

```python
# 获取鼠标当前位置
current_mouse_x, current_mouse_y = pyautogui.position()
print(f'Current mouse position: ({current_mouse_x}, {current_mouse_y})')

# 获取屏幕分辨率
screen_width, screen_height = pyautogui.size()
print(f'Screen resolution: {screen_width}x{screen_height}')
```

5.防错功能

```python
# 启用 failsafe 功能，当鼠标移到屏幕左上角时，程序会抛出异常并停止
pyautogui.FAILSAFE = True

try:
    # 模拟自动操作
    pyautogui.moveTo(0, 0)
    time.sleep(1)
    pyautogui.click()
except pyautogui.FailSafeException:
    print("Failsafe triggered! Program stopped.")
```
